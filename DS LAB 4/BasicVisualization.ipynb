```python
#Importing Necessary Files
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Reading the Data
data= pd.read_csv("car_data.csv")
```

```python
#Bar chart
#A bar chart is used when we want to compare metric values across different subgroups of the data. 
#If we have a greater number of groups, a bar chart is preferred over a column chart.
sns.barplot(x = data['Transmission'],y =data['Kms_Driven'],data = data,palette = "Blues",hue='Transmission')
#Adding the aesthetics
plt.title('Bar Plot')
plt.xlabel('Transmission Type')
plt.ylabel('Kms_Driven') 
# Show the plot
plt.show()
```

```python
#A histogram is a classic visualization tool that represents the distribution 
#of one or more variables by counting the number of observations that fall within disrete bins.

sns.histplot(x= data['Kms_Driven'],data = data,palette = "Blues",bins='auto')
#Adding the aesthetics
plt.title('Histogram')
#plt.y.label('Transmission Type')
plt.xlabel('Kms_Driven') 
# Show the plot
plt.show()
```

```python
#Creating the scatter plot using matplotlib
plt.scatter(data['Selling_Price'],data['Present_Price'],alpha=0.7 )
#Adding the aesthetics
plt.title('Scatter Plot')
plt.xlabel('Selling_Price')
plt.ylabel('Present_Price') 
#Show the plot
plt.show()
```

```python
#Line chart using Matplotlib
plt.plot(data['Selling_Price'],data['Present_Price'],alpha=0.5 )
#Adding the aesthetics
plt.title('Line Plot')
plt.xlabel('Selling_Price')
plt.ylabel('Present_Price') 
#Show the plot
plt.show()
```

```python
#Pie charts can be used to identify proportions of the different components in a given whole.
#Pie chart using Matplotlib
def label_function(val):
    return f'{val / 100 * len(data):.0f}\n{val:.0f}%'

X=data.groupby('Fuel_Type').size()
plt.pie(X, labels = None,autopct=label_function )
#Adding the aesthetics
plt.title('Pie Chart on Fuel_Type')
#Show the plot
plt.show()

Y=data.groupby('Transmission').size()
plt.pie(Y, labels = None,autopct=label_function )
#Adding the aesthetics
plt.title('Pie Chart on Transmission')
#Show the plot
plt.show()
```

```python
#Area chart: Area charts are used to track changes over time for one or more groups. 
#Area graphs are preferred over line charts when we want to capture the changes over time for more than 1 group.
#Reading the dataset
x=range(1,6)
y=[ [1,4,6,8,9], [2,2,7,10,12], [2,8,5,10,6] ]
#Creating the area chart 
ax = plt.gca()
ax.stackplot(x, y, labels=['A','B','C'],alpha=0.7)
#Adding the aesthetics
plt.legend(loc='upper left')
plt.title('area chart ')
plt.xlabel('X ')
plt.ylabel('Y') 
#Show the plot
plt.show()
```

```python
#Area chart using Seaborn
# Data
years_of_experience =[1,2,3]
salary=[ [6,8,10], [4,5,9], [3,5,7] ]
# Plot
plt.stackplot(years_of_experience,salary, labels=['Company A','Company B','Company C'])
plt.legend(loc='upper left')
#Adding the aesthetics
plt.title('Chart title')
plt.xlabel('X axis title')
plt.ylabel('Y axis title') 
# Show the plot
plt.show()
```

```python
#Column histogram using seaborn
#Column histograms are used to observe the distribution for a single variable with few data points.
#Techniques for distribution visualization can provide quick answers to many important questions. 
#What range do the observations cover? What is their central tendency? 
#Are they heavily skewed in one direction? Is there evidence for bimodality? Are there significant outliers? 
#kde=Kernel density estimation
#Plotting bar Distplot
#sns.distplot(data['Kms_Driven'],kde=True, color='blue')
sns.distplot(data['Kms_Driven'],kde=False, color='blue')
plt.title('Distplot')
plt.xlabel('Kms_Driven')
#plt.ylabel('Y axis title') 
# Show the plot
plt.show()
```

```python
#kde plot
sns.displot(data, x=data['Kms_Driven'],kind="kde")
```

```python
#Box plot
#A box plot is used to show the shape of the distribution, its central value, and its variability.
# facilitates comparisons between variables or across levels of a categorical variable. 
#Creating boxplots
sns.boxplot(x=data['Transmission'], y=data['Selling_Price'],data=data, palette="Set3")
#Adding the aesthetics
plt.title('Box Plot')
plt.xlabel('Transmission')
plt.ylabel('Selling_Price') 
# Show the plot
plt.show()



```
