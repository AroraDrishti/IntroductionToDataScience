Bar chart: 
A bar chart is used when we want to compare metric values across different subgroups of the data. If we have a greater number of groups, a bar chart is preferred over a column chart.

```python
plt.bar(x = df['maritalstatus'], height = 100)

#Adding the aesthetics
plt.title('Bar Plot')
plt.xlabel('Marital Status')
plt.ylabel('Frequency') 

#Show the plot
plt.show()
```

Histogram: A histogram is a classic visualization tool that represents the distribution of one or more variables by counting the number of observations that fall within disrete bins.

```python
plt.hist(df['hoursperweek'],bins=None,orientation='vertical')
#Adding the aesthetics
plt.title('Histogram')
plt.ylabel('Count')
plt.xlabel('Hours per week') 
#Show the plot
plt.show()
```

Scatter plot: A scatter diagram is a graphical presentation of the relationship between two quantitative variables.

```python
plt.scatter(df['age'],df['hoursperweek'],s=5,alpha=0.2,linewidths=1)
#Adding the aesthetics
plt.title('Scatter Plot')
plt.xlabel('Age')
plt.ylabel('Hours per week') 
#Show the plot
plt.show()
```

Line Chart: It is a type of chart used to show change of one quantitative data variable over another variable.

```python
plt.plot(df['age'],df['hoursperweek'],'g',alpha=0.5, linewidth=0.2,markersize=1)
#Adding the aesthetics
plt.title('Line Plot')
plt.xlabel('Age')
plt.ylabel('Hours per week') 
#Show the plot
plt.show()
```

Pie chart: It is can be used to identify proportions of the different components in a given whole.

```python
def label_function(val):
    return f'{val / 100 * len(df):.0f}\n{val:.0f}%'

X=df.groupby('race').size()
plt.pie(X, labels = None,radius=3,autopct=label_function)

#Adding the aesthetics
plt.title('Pie Chart grouped by Job Type')
#Show the plot
plt.show()
```

Area chart: Area charts are used to track changes over time for one or more groups. 

```python
X=[1,2]
Y=[[10,20],[30,40],[50,60],[70,80]]

ax = plt.gca()
ax.stackplot(X,Y,labels=['A','B','C','D'],colors='rgyb')

#Adding the aesthetics
plt.legend(loc='upper left')
plt.title('area chart ')
plt.xlabel('X ')
plt.ylabel('Y') 
#Show the plot
plt.show()
```

Box plot: A box plot is used to show the shape of the distribution, its central value, and its variability. It facilitates comparisons between variables or across levels of a categorical variable. 

```python
plt.boxplot(x=df['age'])

#Adding the aesthetics
plt.title('Box Plot')
plt.ylabel('Age') 
# Show the plot
plt.show()
```
