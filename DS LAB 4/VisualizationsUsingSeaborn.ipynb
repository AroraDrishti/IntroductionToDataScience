Importing libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

Importing dataset

```python
df = pd.read_csv('income.csv')
```

```python
df.columns
```

```python
df.dtypes
```

Bar chart: A bar chart is used when we want to compare metric values across different subgroups of the data. If we have a greater number of groups, a bar chart is preferred over a column chart.

```python
bar = sns.barplot(x=df['JobType'],y=df['hoursperweek'],data=df,palette="pastel")

# Adding the aesthetics
bar.set_title('Bar Plot')
bar.set_xlabel('Job Type')
bar.set_ylabel('Hours per week')
```

Histogram: A histogram is a classic visualization tool that represents the distribution of one or more variables by counting the number of observations that fall within disrete bins.

```python
histogram = sns.histplot(x=df['hoursperweek'],data=df,color='b',bins=25)

# Adding the aesthetics
histogram.set_title('Histogram')
histogram.set_xlabel('Hours per week')
histogram.set_ylabel('Count')
```

```python
histogram2 = sns.histplot(x=df['age'],data=df,color='r',bins=25)

# Adding the aesthetics
histogram2.set_title('Histogram')
histogram2.set_xlabel('Age')
histogram2.set_ylabel('Count')
```

Scatter plot: A scatter diagram is a graphical presentation of the relationship between two quantitative variables.

```python
scatterplot = sns.scatterplot(x=df['age'],y=df['hoursperweek'],size=1,data=df,color='g',alpha=0.5)

# Adding the aesthetics
scatterplot.set_title('Scatter Plot')
scatterplot.set_xlabel('Age')
scatterplot.set_ylabel('Hours per week')
```

Line Chart: It is a type of chart used to show change of one quantitative data variable over another variable.

```python
lineplot = sns.lineplot(x=df['age'],y=df['hoursperweek'],size=1,color='c')

# Adding the aesthetics
lineplot.set_title('Line Plot')
lineplot.set_xlabel('Age')
lineplot.set_ylabel('Hours per week')
```

Distribution Plot/ Column histogram: It is used to observe the distribution for a single variable with few data points.

```python
distributionplot = sns.distplot(df['hoursperweek'],bins=10,hist=True,kde=True, color='red')

# Adding the aesthetics
distributionplot.set_title('Distribution Plot')
distributionplot.set_xlabel('Hours per week')
distributionplot.set_ylabel('Frequency')
```

Box plot: A box plot is used to show the shape of the distribution, its central value, and its variability. It facilitates comparisons between variables or across levels of a categorical variable.

```python
boxplot = sns.boxplot(x=df['age'],data=df,palette="Set3")

# Adding the aesthetics
boxplot.set_title('Boxplot')
boxplot.set_xlabel('Age')
```

```python
boxplot2 = sns.boxplot(x=df['JobType'],y=df['age'],data=df,palette="husl")

# Adding the aesthetics
boxplot2.set_title('Boxplot')
boxplot2.set_xlabel('Job Type')
boxplot2.set_ylabel('Age')
```
